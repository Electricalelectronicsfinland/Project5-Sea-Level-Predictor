## 2. Complete Solution (sea_level_predictor.py)

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import linregress
import numpy as np
from matplotlib.ticker import MultipleLocator

#url= "https://raw.githubusercontent.com/Electricalelectronicsfinland/Project5-Sea-Level-Predictor/refs/heads/main/mati-sea-level.csv"

url="https://raw.githubusercontent.com/Electricalelectronicsfinland/Project5-Sea-Level-Predictor/refs/heads/main/espa-sea-level.csv"
df = pd.read_csv(url)

# Configure global plot settings
plt.style.use('seaborn-darkgrid')
plt.rcParams['figure.figsize'] = (14, 7)
plt.rcParams['axes.labelsize'] = 12
plt.rcParams['axes.titlesize'] = 14
###########################################################
def draw_plot():
    # 1. Import and prepare data
    df = pd.read_csv('epa-sea-level.csv') 

###########################################################
    # 2. Create figure with enhanced layout
    fig, ax = plt.subplots()
    
    # 3. Scatter plot with custom markers
    sns.scatterplot(
        x='Year', 
        y='CSIRO Adjusted Sea Level', 
        data=df, 
        color='#1f77b4',
        alpha=0.7,
        edgecolor='w',
        linewidth=0.5,
        s=80,
        ax=ax
    )
    
    # 4. First regression (1880-2050)
    res1 = linregress(df['Year'], df['CSIRO Adjusted Sea Level'])
    x_pred1 = np.arange(1880, 2051)
    y_pred1 = res1.intercept + res1.slope * x_pred1
    ax.plot(x_pred1, y_pred1, 
            color='#d62728', 
            linewidth=2.5,
            label=f'1880-2013 Trend: {res1.slope*100:.2f} in/century')
    
    # 5. Second regression (2000-2050)
    df_recent = df[df['Year'] >= 2000]
    res2 = linregress(df_recent['Year'], df_recent['CSIRO Adjusted Sea Level'])
    x_pred2 = np.arange(2000, 2051)
    y_pred2 = res2.intercept + res2.slope * x_pred2
    ax.plot(x_pred2, y_pred2, 
            color='#2ca02c', 
            linestyle='--',
            linewidth=2.5,
            label=f'2000-2013 Trend: {res2.slope*100:.2f} in/century')
    
    # 6. Add 2050 prediction markers
    ax.scatter(2050, y_pred1[-1], color='#d62728', s=100, zorder=5)
    ax.scatter(2050, y_pred2[-1], color='#2ca02c', s=100, zorder=5)
    
    # 7. Enhanced annotations
    ax.text(2050.5, y_pred1[-1], f'{y_pred1[-1]:.1f} in', va='center')
    ax.text(2050.5, y_pred2[-1], f'{y_pred2[-1]:.1f} in', va='center')
    
    # 8. Formatting
    ax.set(
        xlabel='Year',
        ylabel='Sea Level (inches)',
        title='Rise in Sea Level with Projections to 2050',
        xlim=(1875, 2060),
        ylim=(ax.get_ylim()[0], ax.get_ylim()[1] + 2)
    )
    
    # 9. Custom grid and ticks
    ax.xaxis.set_major_locator(MultipleLocator(20))
    ax.xaxis.set_minor_locator(MultipleLocator(5))
    ax.yaxis.set_major_locator(MultipleLocator(2))
    ax.yaxis.set_minor_locator(MultipleLocator(0.5))
    
    # 10. Professional legend
    ax.legend(
        frameon=True,
        shadow=True,
        borderpad=1,
        title='Regression Lines'
    )
    
    # 11. Save and show
    plt.tight_layout()
    plt.savefig('sea_level_plot.png', dpi=300, bbox_inches='tight')
    return ax

if __name__ == '__main__':
    draw_plot()
    plt.show()
```

```python
!pip install pandas matplotlib seaborn

#################################################################################
# REQUIRED IMPORTS (install with: pip install pandas matplotlib seaborn numpy)
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
##################################################################################
# 1. DATA PREPARATION
def load_and_prepare_data():
    """Load and preprocess sea level data"""
#url="https://raw.githubusercontent.com/Electricalelectronicsfinland/Project5-Sea-Level-Predictor/refs/heads/main/espa-sea-level.csv"
    
    try:
        df = pd.read_csv(url)
        df['Decade'] = (df['Year'] // 10) * 10  # Create decade bins
        return df
    except FileNotFoundError:
        print("Error: Data file not found. Using sample data.")
        np.random.seed(42)
        years = np.arange(1880, 2014)
        sea_levels = np.cumsum(np.random.normal(0.1, 0.5, len(years))) + 5
        return pd.DataFrame({'Year': years, 'CSIRO Adjusted Sea Level': sea_levels})

# 2. PLOT CONFIGURATION
def configure_plot_style():
    """Set professional plot styling"""
    sns.set_theme(
        style="whitegrid",
        palette="muted",
        rc={
            'figure.figsize': (12, 6),
            'axes.titlesize': 14,
            'axes.labelsize': 12,
            'xtick.labelsize': 10,
            'ytick.labelsize': 10,
            'grid.alpha': 0.3
        }
    )

# 3. VIOLIN PLOT FUNCTION
def create_violin_plot(df, save_path='sea_level_violin.png'):
    """
    Creates an optimized violin plot of sea levels by decade
    
    Args:
        df: Pandas DataFrame with 'Decade' and 'CSIRO Adjusted Sea Level' columns
        save_path: Path to save the image
    """
    plt.figure()
    
    # Create the violin plot with error handling
    try:
        ax = sns.violinplot(
            data=df,
            x='Decade',
            y='CSIRO Adjusted Sea Level',
            inner='quartile',  # Show quartile lines
            bw_method=0.2,     # Bandwidth smoothing
            cut=0,             # Show full violin shapes
            linewidth=1        # Edge line thickness
        )
    except Exception as e:
        print(f"Plotting error: {e}")
        return None

    # Enhancements
    ax.set(
        title='Sea Level Distribution by Decade (1880-2013)',
        xlabel='Decade',
        ylabel='Sea Level (inches)'
    )
    
    # Rotate x-axis labels
    plt.xticks(rotation=45, ha='right')
    
    # Add mean markers
    means = df.groupby('Decade')['CSIRO Adjusted Sea Level'].mean()
    for i, decade in enumerate(means.index):
        ax.scatter(i, means[decade], color='red', zorder=10, label='Mean' if i == 0 else "")
    
    # Add legend and save
    ax.legend()
    plt.tight_layout()
    plt.savefig(save_path, dpi=300)
    plt.close()
    print(f"Plot saved to {save_path}")

# MAIN EXECUTION
if __name__ == "__main__":
    # Load data
    sea_level_df = load_and_prepare_data()
    
    # Configure style
    configure_plot_style()
    
    # Generate plot
    create_violin_plot(sea_level_df)
    
    # For Jupyter notebooks:
    try:
        from IPython.display import display, Image
        display(Image(filename='sea_level_violin.png'))
    except ImportError:
        pass
```

```python
# ULTIMATE HEATMAP VISUALIZATION SYSTEM
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.colors import LinearSegmentedColormap
from scipy.cluster import hierarchy
from matplotlib.font_manager import FontProperties
from mpl_toolkits.axes_grid1 import make_axes_locatable
from textwrap import wrap

!pip install pandas matplotlib seaborn scipy numpy

# 1. ENHANCED COLORMAP DESIGN
class HeatmapVisualizer:
    def __init__(self):
        self.pro_font = FontProperties(fname='Roboto-Medium.ttf', size=10) if False else None
        self.cmap = self._create_dynamic_colormap()
        
    def _create_dynamic_colormap(self):
        """Create perceptually uniform diverging colormap"""
        colors = [
            "#313695", "#4575b4", "#74add1", "#abd9e9", "#e0f3f8",
            "#ffffbf", "#fee090", "#fdae61", "#f46d43", "#d73027", "#a50026"
        ]
        return LinearSegmentedColormap.from_list("pro_heat_div", colors, N=512)

# 2. INTELLIGENT DATA PROCESSING
    def _preprocess_data(self, data):
        """Handle all data preprocessing with validation"""
        if isinstance(data, pd.DataFrame):
            df = data.copy()
        else:
            df = pd.DataFrame(data)
            
        # Handle missing values using advanced imputation
        if df.isna().any().any():
            print("Warning: Missing values found, using median imputation")
            df = df.fillna(df.median())
            
        return df

# 3. CLUSTERING ENGINE
    def _cluster_data(self, df, method='ward', optimal_ordering=True):
        """Advanced clustering with multiple linkage methods"""
        try:
            Z = hierarchy.linkage(df, method=method, optimal_ordering=optimal_ordering)
            order = hierarchy.dendrogram(Z, no_plot=True)['leaves']
            return df.iloc[order, :].iloc[:, order]
        except Exception as e:
            print(f"Clustering failed: {e}")
            return df

# 4. ANNOTATION ENGINE
    def _auto_format_annotations(self, val):
        """Smart annotation formatting with dynamic precision"""
        abs_val = abs(val)
        if abs_val == 0:
            return "0"
        elif abs_val < 0.0001:
            return f"{val:.1e}"
        elif abs_val < 0.1:
            return f"{val:.3f}"
        elif abs_val < 1:
            return f"{val:.2f}"
        elif abs_val < 100:
            return f"{val:.1f}"
        elif abs_val < 1e6:
            return f"{int(round(val))}"
        else:
            return f"{val/1e6:.1f}M"

# 5. MAIN RENDERING METHOD
    def render_heatmap(
        self,
        data,
        title="Advanced Heatmap Visualization",
        save_path="ultimate_heatmap_v2.png",
        figsize=(16, 14),
        dpi=400,
        show_values=True,
        cluster=True,
        cluster_method='ward',
        value_range=None,
        cbar_label="Metric Value",
        title_wrap_length=50
    ):
        """
        Generate the ultimate heatmap visualization
        
        Parameters:
            data : Input data (DataFrame or array)
            title : Visualization title
            save_path : Output file path
            figsize : Figure dimensions
            dpi : Output resolution
            show_values : Display cell values
            cluster : Enable clustering
            cluster_method : Linkage method
            value_range : Manual value range (min, max)
            cbar_label : Colorbar label
            title_wrap_length : Title wrapping threshold
        """
        # Data preparation
        df = self._preprocess_data(data)
        
        # Clustering
        if cluster:
            df = self._cluster_data(df, method=cluster_method)
        
        # Create figure with professional layout
        fig, ax = plt.subplots(figsize=figsize)
        fig.subplots_adjust(top=0.92)
        
        # Dynamic value range
        vmin, vmax = (None, None) if value_range is None else value_range
        
        # Generate heatmap
        heatmap = sns.heatmap(
            df,
            cmap=self.cmap,
            center=0 if value_range is None else np.mean(value_range),
            annot=show_values,
            fmt=self._auto_format_annotations,
            annot_kws={
                "size": 8,
                "fontproperties": self.pro_font,
                "color": "black",
                "alpha": 0.8
            },
            linewidths=0.3,
            linecolor='white',
            cbar_kws={
                "shrink": 0.8,
                "label": cbar_label,
                "drawedges": False
            },
            vmin=vmin,
            vmax=vmax,
            ax=ax
        )
        
        # Advanced formatting
        ax.set_title("\n".join(wrap(title, title_wrap_length)), 
                    pad=20, 
                    fontsize=14,
                    fontweight='bold')
        
        # Rotate ticks and adjust alignment
        plt.xticks(
            rotation=45,
            ha='right',
            rotation_mode='anchor',
            fontsize=9
        )
        plt.yticks(fontsize=9)
        
        # Add grid lines
        ax.set_xticks(np.arange(len(df.columns)) + 0.5, minor=True)
        ax.set_yticks(np.arange(len(df.index)) + 0.5, minor=True)
        ax.grid(which="minor", color="white", linestyle='-', linewidth=0.5)
        
        # Add divider for professional colorbar
        divider = make_axes_locatable(ax)
        cax = divider.append_axes("right", size="3%", pad=0.1)
        plt.colorbar(heatmap.collections[0], cax=cax)
        
        # Final polish
        plt.tight_layout()
        
        # Save with metadata
        plt.savefig(
            save_path,
            dpi=dpi,
            bbox_inches='tight',
            transparent=False,
            metadata={
                'Title': title,
                'Author': 'Advanced Heatmap Visualizer',
                'Description': 'Professional heatmap visualization'
            }
        )
        plt.close()
        print(f"Professional heatmap saved to {save_path}")

# EXAMPLE USAGE
if __name__ == "__main__":
    # Initialize visualizer
    visualizer = HeatmapVisualizer()
    
    # Create sample financial data
    np.random.seed(42)
    sectors = ['Tech', 'Finance', 'Healthcare', 'Energy', 'Consumer']
    metrics = ['Revenue', 'Growth', 'Margin', 'ROI', 'Volatility']
    data = pd.DataFrame(
        np.random.randn(5, 5).cumsum(axis=1) * 10 + 50,
        index=sectors,
        columns=metrics
    )
    
    # Generate the ultimate heatmap
    visualizer.render_heatmap(
        data=data,
        title="Sector Performance Metrics Correlation Analysis (2023)",
        save_path="sector_heatmap_pro.png",
        cluster=True,
        cluster_method='average',
        value_range=(-100, 100),
        cbar_label="Performance Score",
        show_values=True
    )
```
